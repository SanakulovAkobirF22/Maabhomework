```python
from datetime import datetime
class Person:
    def __init__(self, name, country, date_of_birth):
        self.name = name
        self.country = country
        self.date_of_birth = datetime.strptime(date_of_birth, '%Y-%m-%d')
    
    def get_age(self):
        today = datetime.today()
        age = today.year - self.date_of_birth.year

        # Adjust if birthday hasn't occured yet this year
        if (today.month, today.day) < (self.date_of_birth.month, self.date_of_birth.day):
            age -= 1
        return age



person1 = Person('Alice', 'USA', '2000-06-15')

print('Name:', person1.name)
print('Country:', person1.country)
print('Age:', person1.get_age())
```

```python
import math

# Base class

class Shape:
    def area(self):
        raise NotImplementedError("This method nust be overriden in a subclass.")
    
    def perimeter(self):
        raise NotImplementedError("This method must be overriden in a subclass")
    

# Subclass: Circle

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius**2
    
    def perimeter(self):
        return 2*math.pi*self.radius
    

# Subclass: Square

class square(Shape):
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side ** 2
    
    def perimeter(self):
        return 4*self.side
    


# Subclass: Triangle (for simplicity, assume it's an equilateral triangle)
class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def perimeter(self):
        return self.a + self.b + self.c
    
    def area(self):
        # Heron's formula
        p = self.perimeter()/2

        return math.sqrt(p * (p - self.a)*(p - self.b)*(p - self.c))
    
    
c = Circle(5)
print("Circle -> Area:", round(c.area(), 2), "| Perimeter:", round(c.perimeter(), 2))

s = square(4)
print("square -> Area:", s.area(), "| Perimeter:", s.perimeter())

p = Triangle(3, 4, 5)
print("Triangle -> Area:", round(p.area(), 2), "| Perimeter:", p.perimeter())

    
```

```python
# Node class
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Binary Search Tree class
class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Insert method
    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert_recursive(self.root, value)

    def _insert_recursive(self, current, value):
        if value < current.value:
            if current.left is None:
                current.left = Node(value)
            else:
                self._insert_recursive(current.left, value)
        elif value > current.value:
            if current.right is None:
                current.right = Node(value)
            else:
                self._insert_recursive(current.right, value)
        # If value == current.value, we ignore duplicates (optional)

    # Search method
    def search(self, value):
        return self._search_recursive(self.root, value)

    def _search_recursive(self, current, value):
        if current is None:
            return False
        if current.value == value:
            return True
        elif value < current.value:
            return self._search_recursive(current.left, value)
        else:
            return self._search_recursive(current.right, value)



bst = BinarySearchTree()
bst.insert(50)
bst.insert(30)
bst.insert(70)
bst.insert(20)
bst.insert(40)
bst.insert(60)
bst.insert(80)

print("Searching for 60:", bst.search(60))  # True
print("Searching for 25:", bst.search(25))  # False
```

```python
class Stack:
    """
    A compact class representing a Stack (LIFO).
    """
    def __init__(self):
        """Initializes an empty stack."""
        self.items = []

    def push(self, element):
        """Adds an element to the top."""
        self.items.append(element)

    def pop(self):
        """Removes and returns the top element."""
        if not self.items: # More concise check for empty list
            raise IndexError("pop from empty stack")
        return self.items.pop()

    def is_empty(self):
        """Checks if the stack is empty."""
        return not self.items # Concise check: empty list is Falsy

    def peek(self):
        """Returns the top element without removing it."""
        if not self.items:
            raise IndexError("peek from empty stack")
        return self.items[-1]

    def size(self):
        """Returns the number of elements."""
        return len(self.items)

# --- How to use the shorter Stack class ---

if __name__ == "__main__":
    my_stack = Stack()

    print(f"Stack empty: {my_stack.is_empty()}")
    my_stack.push(1)
    my_stack.push(2)
    my_stack.push(3)
    print(f"Stack size: {my_stack.size()}")
    print(f"Top element: {my_stack.peek()}")
    print(f"Popped: {my_stack.pop()}")
    print(f"Stack size: {my_stack.size()}")

    while not my_stack.is_empty():
        print(f"Popped: {my_stack.pop()}")

    print(f"Stack empty: {my_stack.is_empty()}")

    try:
        my_stack.pop()
    except IndexError as e:
        print(f"Error: {e}")
```

```python
class Node:
    """Represents a node in a linked list."""
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    """
    A compact class representing a linked list.
    Includes methods for display, insertion, and deletion.
    """
    def __init__(self):
        self.head = None

    def display(self):
        """Displays all data in the linked list."""
        current = self.head
        if not current:
            print("List is empty.")
            return

        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements) + " -> None")

    def insert_at_beginning(self, data):
        """Inserts a new node at the beginning."""
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        # print(f"Inserted {data} at beginning.") # Removed for brevity

    def insert_at_end(self, data):
        """Inserts a new node at the end."""
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
        # print(f"Inserted {data} at end.") # Removed for brevity

    def insert_after_node(self, prev_node_data, new_data):
        """Inserts a new node after a specific node."""
        current = self.head
        while current and current.data != prev_node_data:
            current = current.next

        if not current:
            print(f"Node with data '{prev_node_data}' not found.")
            return

        new_node = Node(new_data)
        new_node.next = current.next
        current.next = new_node
        # print(f"Inserted {new_data} after {prev_node_data}.") # Removed for brevity

    def delete_node(self, key):
        """Deletes the first node with the given key."""
        current = self.head

        if current and current.data == key:
            self.head = current.next
            # print(f"Deleted {key} from beginning.") # Removed for brevity
            return

        prev = None
        while current and current.data != key:
            prev = current
            current = current.next

        if not current:
            print(f"Node with data '{key}' not found.")
            return

        prev.next = current.next
        # print(f"Deleted {key}.") # Removed for brevity


# --- Usage Example (with fewer print statements in the class) ---

if __name__ == "__main__":
    my_list = LinkedList()

    print("Initial list:")
    my_list.display()

    print("\nInserting elements:")
    my_list.insert_at_beginning(5)
    my_list.insert_at_beginning(10)
    my_list.insert_at_end(15)
    my_list.insert_at_end(20)
    my_list.insert_after_node(10, 7)
    my_list.display() # Should be: 10 -> 7 -> 5 -> 15 -> 20 -> None

    print("\nDeleting elements:")
    my_list.delete_node(10) # Delete head
    my_list.display() # Should be: 7 -> 5 -> 15 -> 20 -> None

    my_list.delete_node(15) # Delete middle
    my_list.display() # Should be: 7 -> 5 -> 20 -> None

    my_list.delete_node(20) # Delete last
    my_list.display() # Should be: 7 -> 5 -> None

    my_list.delete_node(99) # Not found
    my_list.display() # Should be unchanged

    my_list.delete_node(5) # Delete last
    my_list.display() # Should be: 7 -> None

    my_list.delete_node(7) # Delete last
    my_list.display() # Should be empty

    print("\nAttempting to delete from empty list:")
    my_list.delete_node(1) # Delete from empty
    my_list.display()
```

```python
class ShoppingCart:
    """
    A class representing a shopping cart.
    Allows adding, removing items, and calculating the total price.
    """
    def __init__(self):
        
        self.items = []

    def add_item(self, item_name, price):
        
        item = {'name': item_name, 'price': price}
        self.items.append(item)
        print(f"Added '{item_name}' (Price: ${price:.2f}) to the cart.")

    def remove_item(self, item_name):
        
        found = False
        # Iterate through the list to find and remove the item
        for i, item in enumerate(self.items):
            if item['name'] == item_name:
                self.items.pop(i) # Remove the item by its index
                found = True
                print(f"Removed '{item_name}' from the cart.")
                break # Exit loop after finding and removing the first occurrence
        
        if not found:
            print(f"'{item_name}' not found in the cart.")

    def calculate_total(self):
        

        total = 0.0
        for item in self.items:
            total += item['price']
        return total

    def display_cart(self):
        """
        Displays all items currently in the cart and the total price.
        """
        if not self.items:
            print("Your shopping cart is empty.")
            return

        print("\n--- Your Shopping Cart ---")
        for item in self.items:
            print(f"- {item['name']}: ${item['price']:.2f}")
        
        total = self.calculate_total()
        print(f"--------------------------")
        print(f"Total: ${total:.2f}")
        print("--------------------------")


# --- How to use the ShoppingCart class ---

if __name__ == "__main__":
    my_cart = ShoppingCart()

    my_cart.display_cart() # Should show empty cart

    print("\n--- Adding Items ---")
    my_cart.add_item("Laptop", 1200.00)
    my_cart.add_item("Mouse", 25.50)
    my_cart.add_item("Keyboard", 75.00)
    my_cart.add_item("Laptop", 1200.00) # Adding another laptop

    my_cart.display_cart()

    print("\n--- Removing Items ---")
    my_cart.remove_item("Mouse")
    my_cart.remove_item("Headphones") # Item not in cart

    my_cart.display_cart()

    print("\n--- Final Total ---")
    final_total = my_cart.calculate_total()
    print(f"The final total price of items in your cart is: ${final_total:.2f}")

    print("\n--- Emptying Cart ---")
    my_cart.remove_item("Laptop")
    my_cart.remove_item("Laptop")
    my_cart.remove_item("Keyboard")
    my_cart.display_cart() # Should be empty
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, element):
        self.items.append(element)
        print(f"Pushed: {element}")

    def pop(self):
        if not self.items:  # Checks if the list is empty (Pythonic way)
            raise IndexError("Cannot pop from an empty stack.")
        popped_item = self.items.pop()
        print(f"Popped: {popped_item}")
        return popped_item

    def is_empty(self):
        return not self.items # Returns True if self.items is empty, False otherwise

    def display(self):
        if self.is_empty():
            print("Stack is empty.")
            return

        print("\n--- Stack (Top to Bottom) ---")
        # Iterate in reverse to show from top to bottom
        for item in reversed(self.items):
            print(f"| {item} |")
        print("-------") # Base of the stack

# --- How to use the Stack class ---

if __name__ == "__main__":
    my_stack = Stack()

    my_stack.display()

    print("\n--- Pushing elements ---")
    my_stack.push(10)
    my_stack.push(20)
    my_stack.push(30)
    my_stack.display()

    print("\n--- Popping elements ---")
    first_popped = my_stack.pop()
    second_popped = my_stack.pop()
    my_stack.display()

    print("\n--- Pushing more elements ---")
    my_stack.push(40)
    my_stack.push(50)
    my_stack.display()

    print("\n--- Emptying the stack ---")
    while not my_stack.is_empty():
        my_stack.pop()
    my_stack.display()

    print("\n--- Attempting to pop from empty stack ---")
    try:
        my_stack.pop()
    except IndexError as e:
        print(f"Error: {e}")
```

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, element):
        self.items.append(element)
        print(f"Enqueued: {element}")

    def dequeue(self):
        if not self.items:
            raise IndexError("Cannot dequeue from an empty queue.")
        
        dequeued_item = self.items.pop(0) 
        print(f"Dequeued: {dequeued_item}")
        return dequeued_item

    def is_empty(self):
        return not self.items

    def display(self):
        if self.is_empty():
            print("Queue is empty.")
            return

        print("\n--- Queue (Front to Rear) ---")
        print(" <- ".join(map(str, self.items)))
        print("-----------------------------")


# --- How to use the Queue class ---

if __name__ == "__main__":
    my_queue = Queue()

    my_queue.display()

    print("\n--- Enqueueing elements ---")
    my_queue.enqueue("Task A")
    my_queue.enqueue("Task B")
    my_queue.enqueue("Task C")
    my_queue.display()

    print("\n--- Dequeueing elements ---")
    first_task = my_queue.dequeue()
    second_task = my_queue.dequeue()
    my_queue.display()

    print("\n--- Enqueueing more elements ---")
    my_queue.enqueue("Task D")
    my_queue.enqueue("Task E")
    my_queue.display()

    print("\n--- Emptying the queue ---")
    while not my_queue.is_empty():
        my_queue.dequeue()
    my_queue.display()

    print("\n--- Attempting to dequeue from empty queue ---")
    try:
        my_queue.dequeue()
    except IndexError as e:
        print(f"Error: {e}")
        
```

```python
class Account:
    """
    Represents a single bank account for a customer.
    """
    def __init__(self, account_number, customer_name, initial_balance=0.0):
        if initial_balance < 0:
            raise ValueError("Initial balance cannot be negative.")
        self.account_number = account_number
        self.customer_name = customer_name
        self._balance = initial_balance  # Use _balance to indicate it's internal

    def get_balance(self):
        """Returns the current balance of the account."""
        return self._balance

    def deposit(self, amount):
        """
        Deposits money into the account.
        Raises ValueError if amount is not positive.
        """
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self._balance += amount
        return True # Indicate success

    def withdraw(self, amount):
        """
        Withdraws money from the account.
        Raises ValueError if amount is not positive or if funds are insufficient.
        """
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if amount > self._balance:
            raise ValueError("Insufficient funds.")
        self._balance -= amount
        return True # Indicate success

    def get_info(self):
        """Returns a string with account details."""
        return (f"Account No: {self.account_number}, "
                f"Customer: {self.customer_name}, "
                f"Balance: ${self._balance:.2f}")


class Bank:
    """
    Manages customer accounts and transactions.
    """
    def __init__(self):
        self.accounts = {}  # Dictionary to store Account objects: {account_number: Account_object}
        self._next_account_number = 1001  # Starting account number

    def create_account(self, customer_name, initial_deposit=0.0):
        """
        Creates a new customer account and adds it to the bank.
        Returns the new account number or None if creation fails.
        """
        try:
            new_account_number = self._next_account_number
            new_account = Account(new_account_number, customer_name, initial_deposit)
            self.accounts[new_account_number] = new_account
            self._next_account_number += 1
            print(f"Account created successfully for {customer_name}. Account No: {new_account_number}")
            return new_account_number
        except ValueError as e:
            print(f"Failed to create account: {e}")
            return None

    def _get_account(self, account_number):
        """Helper method to retrieve an account object. Returns None if not found."""
        account = self.accounts.get(account_number)
        if not account:
            print(f"Error: Account {account_number} not found.")
        return account

    def deposit(self, account_number, amount):
        """Deposits money into a specified account."""
        account = self._get_account(account_number)
        if account:
            try:
                account.deposit(amount)
                print(f"Successfully deposited ${amount:.2f} into account {account_number}.")
                return True
            except ValueError as e:
                print(f"Deposit failed for account {account_number}: {e}")
        return False

    def withdraw(self, account_number, amount):
        """Withdraws money from a specified account."""
        account = self._get_account(account_number)
        if account:
            try:
                account.withdraw(amount)
                print(f"Successfully withdrew ${amount:.2f} from account {account_number}.")
                return True
            except ValueError as e:
                print(f"Withdrawal failed for account {account_number}: {e}")
        return False

    def transfer(self, from_account_num, to_account_num, amount):
        """Transfers money between two accounts."""
        if amount <= 0:
            print("Transfer amount must be positive.")
            return False

        from_account = self._get_account(from_account_num)
        to_account = self._get_account(to_account_num)

        if not from_account or not to_account:
            print("Transfer failed: One or both accounts not found.")
            return False

        if from_account.get_balance() < amount:
            print(f"Transfer failed: Insufficient funds in account {from_account_num}.")
            return False

        try:
            # Perform withdrawal first
            from_account.withdraw(amount)
            # If withdrawal is successful, then perform deposit
            to_account.deposit(amount)
            print(f"Successfully transferred ${amount:.2f} from {from_account_num} to {to_account_num}.")
            return True
        except ValueError as e:
            # This catch is mostly for safety, as conditions are checked above
            print(f"Transfer operation failed: {e}")
            return False

    def get_account_balance(self, account_number):
        """Returns the balance of a specific account."""
        account = self._get_account(account_number)
        if account:
            print(f"Balance for account {account_number}: ${account.get_balance():.2f}")
            return account.get_balance()
        return None

    def display_all_accounts(self):
        """Displays details for all accounts in the bank."""
        if not self.accounts:
            print("No accounts in the bank yet.")
            return

        print("\n--- All Bank Accounts ---")
        for acc_num, account in self.accounts.items():
            print(account.get_info())
        print("------------------------")


# --- How to use the Bank and Account classes ---

if __name__ == "__main__":
    my_bank = Bank()

    # 1. Create Accounts
    print("--- Creating Accounts ---")
    acc1_num = my_bank.create_account("Alice Smith", 500.00)
    acc2_num = my_bank.create_account("Bob Johnson", 1200.50)
    acc3_num = my_bank.create_account("Charlie Brown") # Default initial balance 0

    my_bank.display_all_accounts()

    # 2. Deposit money
    print("\n--- Depositing Money ---")
    my_bank.deposit(acc1_num, 200.00)
    my_bank.deposit(acc2_num, 50.00)
    my_bank.deposit(9999, 100.00) # Non-existent account
    my_bank.deposit(acc1_num, -50.00) # Invalid deposit amount

    my_bank.display_all_accounts()

    # 3. Withdraw money
    print("\n--- Withdrawing Money ---")
    my_bank.withdraw(acc1_num, 100.00)
    my_bank.withdraw(acc2_num, 1500.00) # Insufficient funds
    my_bank.withdraw(acc3_num, 10.00) # Withdraw from zero balance
    my_bank.withdraw(acc1_num, -20.00) # Invalid withdrawal amount

    my_bank.display_all_accounts()

    # 4. Get Account Balances
    print("\n--- Checking Balances ---")
    my_bank.get_account_balance(acc1_num)
    my_bank.get_account_balance(acc2_num)
    my_bank.get_account_balance(9999) # Non-existent account

    # 5. Transfer money
    print("\n--- Transferring Money ---")
    my_bank.transfer(acc1_num, acc2_num, 150.00) # Successful transfer
    my_bank.transfer(acc2_num, acc1_num, 5000.00) # Insufficient funds
    my_bank.transfer(acc1_num, 9999, 50.00) # Non-existent destination
    my_bank.transfer(9999, acc1_num, 50.00) # Non-existent source

    my_bank.display_all_accounts()

    # Show final balances
    print("\n--- Final Balances ---")
    my_bank.get_account_balance(acc1_num)
    my_bank.get_account_balance(acc2_num)
    my_bank.get_account_balance(acc3_num)
```
