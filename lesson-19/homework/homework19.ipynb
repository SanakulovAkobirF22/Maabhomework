```python
import pandas as pd
import numpy as np
```

```python
sales_df = pd.read_csv("https://raw.githubusercontent.com/IslomovSH/python-homework/refs/heads/main/lesson-19/homework/task/sales_data.csv")
sales_df.head()
```

```python
result = sales_df.groupby('Category').agg({'Quantity': ['sum', 'max'], 'Price':'mean'})
result 
```

```python
grouped = sales_df.groupby(['Category', 'Product'], as_index=False)['Quantity'].sum()

top_selling = grouped.sort_values(['Category', 'Quantity'], ascending=[False, True]).drop_duplicates(subset='Category')
top_selling
```

```python
sales_df['TotalSales'] = sales_df.Quantity * sales_df.Price

daily_sales = sales_df.groupby('Date', as_index=False)['TotalSales'].sum()

top_day = daily_sales.loc[daily_sales['TotalSales'].idxmax()]

top_day
```

```python
cus_df = pd.read_csv("https://raw.githubusercontent.com/IslomovSH/python-homework/refs/heads/main/lesson-19/homework/task/customer_orders.csv")
cus_df
```

```python
grp = cus_df.groupby('CustomerID', as_index=False)['Quantity'].sum()
grp[grp.Quantity < 20]
```

```python
avgl = cus_df.groupby('CustomerID', as_index=False)['Price'].mean()
avgl[avgl.Price > 120]
```

```python
pro_sum = cus_df.groupby('Product', as_index=False).agg({
    'Quantity':'sum',
    'Price': lambda p: (p * cus_df.loc[p.index, 'Quantity']).sum()
})

filt = pro_sum[pro_sum['Quantity'] >= 5]
filt
```
