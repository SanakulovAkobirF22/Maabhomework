```python
# math_operations.py

def add(a, b):
    """
    Эта функция складывает два числа и возвращает их сумму.
    Аргументы:
        a (int/float): Первое число.
        b (int/float): Второе число.
    Возвращает:
        int/float: Сумма a и b.
    """
    return a + b

def subtract(a, b):
    """
    Эта функция вычитает второе число из первого и возвращает результат.
    Аргументы:
        a (int/float): Число, из которого вычитают.
        b (int/float): Число, которое вычитают.
    Возвращает:
        int/float: Разность a и b.
    """
    return a - b

def multiply(a, b):
    """
    Эта функция умножает два числа и возвращает их произведение.
    Аргументы:
        a (int/float): Первое число.
        b (int/float): Второе число.
    Возвращает:
        int/float: Произведение a и b.
    """
    return a * b

def divide(a, b):
    """
    Эта функция делит первое число на второе и возвращает результат.
    В случае деления на ноль, она возвращает сообщение об ошибке.
    Аргументы:
        a (int/float): Делимое.
        b (int/float): Делитель.
    Возвращает:
        int/float или str: Частное a и b, или сообщение об ошибке.
    """
    if b == 0:
        return "Ошибка: Деление на ноль!"
    return a / b

# Пример использования (этот блок кода будет выполняться только при прямом запуске файла)
if __name__ == "__main__":
    print("Тестируем функции из math_operations.py:")
    print(f"5 + 3 = {add(5, 3)}")
    print(f"10 - 4 = {subtract(10, 4)}")
    print(f"6 * 7 = {multiply(6, 7)}")
    print(f"10 / 2 = {divide(10, 2)}")
    print(f"7 / 0 = {divide(7, 0)}")
```

```python
# string_utils.py

def reverse_string(s):
    """
    Эта функция переворачивает заданную строку.

    Аргументы:
        s (str): Входная строка.

    Возвращает:
        str: Перевернутая строка.
    """
    return s[::-1]

def count_vowels(s):
    """
    Эта функция подсчитывает количество гласных букв (a, e, i, o, u)
    в заданной строке, без учета регистра.

    Аргументы:
        s (str): Входная строка.

    Возвращает:
        int: Количество гласных букв.
    """
    vowels = "aeiouAEIOU" # Определяем все гласные буквы (строчные и заглавные)
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

# Пример использования (этот блок кода будет выполняться только при прямом запуске файла)
if __name__ == '__main__':
    print("Тестируем функции из string_utils.py:")

    test_string_1 = "Hello Python"
    print(f"Оригинальная строка: '{test_string_1}'")
    print(f"Перевернутая строка: '{reverse_string(test_string_1)}'")
    print(f"Количество гласных: {count_vowels(test_string_1)}")

    print("-" * 20)

    test_string_2 = "Programming is fun"
    print(f"Оригинальная строка: '{test_string_2}'")
    print(f"Перевернутая строка: '{reverse_string(test_string_2)}'")
    print(f"Количество гласных: {count_vowels(test_string_2)}")

    print("-" * 20)

    test_string_3 = "AEIOU"
    print(f"Оригинальная строка: '{test_string_3}'")
    print(f"Перевернутая строка: '{reverse_string(test_string_3)}'")
    print(f"Количество гласных: {count_vowels(test_string_3)}")
```

```python
# geometry/circle.py

import math

def calculate_area(radius):
    """
    Calculates the area of a circle given its radius.

    Args:
        radius (float or int): The radius of the circle.

    Returns:
        float: The area of the circle.

    Raises:
        ValueError: If the radius is negative.
    """
    if radius < 0:
        raise ValueError("Radius cannot be negative.")
    return math.pi * (radius ** 2)

def calculate_circumference(radius):
    """
    Calculates the circumference (perimeter) of a circle given its radius.

    Args:
        radius (float or int): The radius of the circle.

    Returns:
        float: The circumference of the circle.

    Raises:
        ValueError: If the radius is negative.
    """
    if radius < 0:
        raise ValueError("Radius cannot be negative.")
    return 2 * math.pi * radius

# Example usage (this block of code will only run when the module is executed directly)
if __name__ == '__main__':
    print("Testing functions from geometry/circle.py:")

    r1 = 5
    print(f"Circle with radius {r1}:")
    print(f"  Area: {calculate_area(r1):.2f}") # Formats to two decimal places
    print(f"  Circumference: {calculate_circumference(r1):.2f}")

    print("-" * 30)

    r2 = 10.5
    print(f"Circle with radius {r2}:")
    print(f"  Area: {calculate_area(r2):.2f}")
    print(f"  Circumference: {calculate_circumference(r2):.2f}")

    print("-" * 30)

    try:
        calculate_area(-1)
    except ValueError as e:
        print(f"Error calculating with negative radius: {e}")
```

```python
# file_operations/file_reader.py

def read_file(file_path):
    """
    Reads the entire content of a text file.

    Args:
        file_path (str): The path to the file to be read.

    Returns:
        str: The content of the file, or an error message if the file cannot be read.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return f"Error: File not found at '{file_path}'"
    except Exception as e:
        return f"An error occurred while reading the file: {e}"

# Example usage (runs only when this module is executed directly)
if __name__ == '__main__':
    # Create a dummy file for testing
    test_file_name = "test_read.txt"
    with open(test_file_name, 'w', encoding='utf-8') as f:
        f.write("This is a test line.\n")
        f.write("Another line for reading.\n")

    print(f"--- Testing read_file function with '{test_file_name}' ---")
    read_content = read_file(test_file_name)
    print("File Content:")
    print(read_content)

    print("\n--- Testing read_file function with a non-existent file ---")
    non_existent_file = "non_existent.txt"
    error_message = read_file(non_existent_file)
    print(error_message)
```

```python
# file_operations/file_writer.py

def write_file(file_path, content):
    """
    Writes the given content to a text file.
    If the file exists, its content will be overwritten.
    If the file does not exist, it will be created.

    Args:
        file_path (str): The path to the file where content will be written.
        content (str): The string content to write to the file.

    Returns:
        bool: True if writing was successful, False otherwise.
    """
    try:
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(content)
        return True
    except Exception as e:
        print(f"An error occurred while writing to the file: {e}")
        return False

# Example usage (runs only when this module is executed directly)
if __name__ == '__main__':
    test_write_file_name = "test_write.txt"
    test_content = "Hello, this is some content written to a file.\n" \
                   "It's great to work with Python file operations!"

    print(f"--- Testing write_file function to '{test_write_file_name}' ---")
    if write_file(test_write_file_name, test_content):
        print("Content successfully written to the file.")
        # Verify by reading it back
        from .file_reader import read_file # Relative import for testing within package
        read_back_content = read_file(test_write_file_name)
        print("\nContent read back from file:")
        print(read_back_content)
    else:
        print("Failed to write content to the file.")

    print("\n--- Testing write_file with different content ---")
    if write_file(test_write_file_name, "This is new content, overwriting the old."):
        print("New content successfully written.")
        from .file_reader import read_file
        read_back_content = read_file(test_write_file_name)
        print("\nContent read back from file:")
        print(read_back_content)
```
