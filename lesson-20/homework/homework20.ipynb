```python
import sqlite3
import pandas as pd


db_path = 'chinook.db'
try:
    
    conn = sqlite3.connect(db_path)

    # --- Step 1: Load all necessary tables into pandas DataFrames ---
    invoice_items_df = pd.read_sql_query("SELECT InvoiceId, TrackId FROM invoice_items", conn)
    tracks_df = pd.read_sql_query("SELECT TrackId, AlbumId FROM tracks", conn)
    invoices_df = pd.read_sql_query("SELECT InvoiceId, CustomerId FROM invoices", conn)
    albums_df = pd.read_sql_query("SELECT AlbumId FROM albums", conn)

    print("Successfully loaded data into pandas DataFrames.")

except sqlite3.Error as e:
    print(f"Database error: {e}")
    # Exit or handle the error appropriately if the script cannot proceed
    exit()
finally:
    if conn:
        conn.close()


album_track_counts = tracks_df.groupby('AlbumId').size().reset_index(name='TotalAlbumTracks')

invoice_details_df = pd.merge(invoice_items_df, tracks_df, on='TrackId')


tracks_bought_per_album = invoice_details_df.groupby(['InvoiceId', 'AlbumId']).size().reset_index(name='TracksBought')


album_completeness_df = pd.merge(tracks_bought_per_album, album_track_counts, on='AlbumId')

album_completeness_df['IsAlbumPurchase'] = album_completeness_df['TracksBought'] == album_completeness_df['TotalAlbumTracks']


invoice_type_df = album_completeness_df.groupby('InvoiceId')['IsAlbumPurchase'].min().reset_index()


customer_purchases_df = pd.merge(invoices_df, invoice_type_df, on='InvoiceId')


customer_summary = customer_purchases_df.groupby('CustomerId')['IsAlbumPurchase'].agg(
    AlbumInvoices='sum',
    TotalInvoices='count'
).reset_index()


customer_summary['TrackInvoices'] = customer_summary['TotalInvoices'] - customer_summary['AlbumInvoices']


track_preferring = 0
album_preferring = 0
mixed = 0

for index, row in customer_summary.iterrows():
    if row['TrackInvoices'] > row['AlbumInvoices']:
        track_preferring += 1
    elif row['AlbumInvoices'] > row['TrackInvoices']:
        album_preferring += 1
    else:
        mixed += 1

total_customers = len(customer_summary)

if total_customers > 0:
    # Calculate final percentages
    percent_track = (track_preferring / total_customers) * 100
    percent_album = (album_preferring / total_customers) * 100
    percent_mixed = (mixed / total_customers) * 100

    # --- Print the final results ---
    print("\nCustomer Purchasing Preference Analysis (Pandas Script)")
    print("-" * 50)
    print(f"Total Customers Analyzed: {total_customers}")
    print(f"Customers Preferring Individual Tracks: {track_preferring} ({percent_track:.2f}%)")
    print(f"Customers Preferring Full Albums: {album_preferring} ({percent_album:.2f}%)")
    print(f"Customers with Equal Behavior: {mixed} ({percent_mixed:.2f}%)")
else:
    print("No customer data to analyze.")
```

```python
db_path = 'chinook.db'
try:
    
    conn = sqlite3.connect(db_path)

    
    invoice_items_df = pd.read_sql_query("SELECT InvoiceId, TrackId FROM invoice_items", conn)
    tracks_df = pd.read_sql_query("SELECT TrackId, AlbumId FROM tracks", conn)
    invoices_df = pd.read_sql_query("SELECT InvoiceId, CustomerId FROM invoices", conn)
    albums_df = pd.read_sql_query("SELECT AlbumId FROM albums", conn)

    print("Successfully loaded data into pandas DataFrames.")

except sqlite3.Error as e:
    print(f"Database error: {e}")
    
    exit()
finally:
    if conn:
        conn.close()


album_track_counts = tracks_df.groupby('AlbumId').size().reset_index(name='TotalAlbumTracks')


invoice_details_df = pd.merge(invoice_items_df, tracks_df, on='TrackId')

tracks_bought_per_album = invoice_details_df.groupby(['InvoiceId', 'AlbumId']).size().reset_index(name='TracksBought')

album_completeness_df = pd.merge(tracks_bought_per_album, album_track_counts, on='AlbumId')
album_completeness_df['IsAlbumPurchase'] = album_completeness_df['TracksBought'] == album_completeness_df['TotalAlbumTracks']


invoice_type_df = album_completeness_df.groupby('InvoiceId')['IsAlbumPurchase'].min().reset_index()


customer_purchases_df = pd.merge(invoices_df, invoice_type_df, on='InvoiceId')


customer_preference = customer_purchases_df.groupby('CustomerId')['IsAlbumPurchase'].min().reset_index()


total_customers = len(customer_preference)

if total_customers > 0:
    # Count customers who have ONLY purchased full albums (where the minimum was True).
    album_only_customers = customer_preference['IsAlbumPurchase'].sum()

    # The rest are customers who have purchased individual tracks at least once.
    individual_track_customers = total_customers - album_only_customers

    # Calculate final percentages
    percent_individual = (individual_track_customers / total_customers) * 100
    percent_album_only = (album_only_customers / total_customers) * 100

    # --- Print the final results ---
    print("\nCustomer Purchasing Preference Analysis (Updated Logic)")
    print("-" * 50)
    print(f"Total Customers Analyzed: {total_customers}")
    print(f"Customers who have purchased individual tracks: {individual_track_customers} ({percent_individual:.2f}%)")
    print(f"Customers who have ONLY purchased full albums: {album_only_customers} ({percent_album_only:.2f}%)")
else:
    print("No customer data to analyze.")
```
